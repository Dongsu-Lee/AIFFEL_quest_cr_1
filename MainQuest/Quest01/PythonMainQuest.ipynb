Q1. 계좌 만들기
===============

```python
import random

class Account:
    
    account_count = 0
    
    def __init__(self, owner, balance):
        self.bank_name = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.generate_account_number()
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []
        
        Account.account_count += 1
    
    def generate_account_number(self):
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))
        return f"{part1}-{part2}-{part3}"
    
    @classmethod
    def get_account_num(cls):
        return cls.account_count
    
    def deposit(self, amount):
        if amount < 1:
            print("입금은 1원 이상만 가능합니다.")
        else:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_history_list.append(amount)
            print(f"{amount}원이 입금되었습니다. {self.owner} 현재 잔액: {self.balance:,}원")
            
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                print(f"이자가 지급되었습니다! {self.owner} 현재 잔액: {self.balance:,}원")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        else:
            self.balance -= amount
            self.withdraw_history_list.append(amount)
            print(f"{amount}원이 출금되었습니다. {self.owner} 현재 잔액: {self.balance:,}원")
    
    def deposit_history(self):
        print("입금 내역:", self.deposit_history_list)
    
    def withdraw_history(self):
        print("출금 내역:", self.withdraw_history_list)
    
    def display_info(self):
        balance_formatted = f"{self.balance:,}원"
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {balance_formatted}")

accounts = [
    Account("카카로트", 500000),
    Account("베지터", 1800000),
    Account("브로리", 300000),
    Account("마인부우", 20000000)
]

for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

target_account = accounts[1]  

target_account.deposit(100000)
target_account.deposit(200000)
target_account.withdraw(150000)
target_account.deposit(50000)
target_account.deposit(100000)
target_account.deposit(250000)

target_account.deposit_history()
target_account.withdraw_history()
```

Q2. RPG 게임 만들기
==================

```python
import random

class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense
    
    def is_alive(self):
        return self.hp > 0
    
    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense)
        self.hp -= actual_damage
        self.hp = max(0, self.hp)  # 체력이 음수가 되지 않도록 처리
        print(f"{self.name}이(가) {actual_damage}만큼의 피해를 입었습니다. 남은 HP: {self.hp}")
    
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}이(가) {target.name}에게 {damage}만큼의 피해를 입힙니다.")
        target.take_damage(damage)

# 예시
character1 = Character("용사", 1, 100, 20, 5)
character2 = Character("슬라임", 1, 50, 10, 2)

character1.attack_target(character2)
character2.attack_target(character1)

print(f"{character1.name} 생존 여부: {character1.is_alive()}")
print(f"{character2.name} 생존 여부: {character2.is_alive()}")
```

```python
# Player 클래스
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0
    
    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) {exp} 경험치를 획득했습니다. 현재 경험치: {self.experience}")
        if self.experience >= 50:
            self.level_up()
    
    def level_up(self):
        self.level += 1
        self.attack += 10
        self.defense += 5
        self.hp += 20
        self.experience = 0
        print(f"{self.name}이(가) 레벨업 했습니다! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}, 체력: {self.hp}")

# Monster 클래스
class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)
        print(f"몬스터 {self.name}이(가) 생성되었습니다. 레벨: {self.level}, 체력: {self.hp}, 공격력: {self.attack}, 방어력: {self.defense}")

# 예시
player = Player("용사")
monster = Monster("슬라임", 3)

player.attack_target(monster)
monster.attack_target(player)

player.gain_experience(30)
player.gain_experience(25)  # 경험치가 50이 넘어 레벨업

player.attack_target(monster)
```

```python
# battle 함수
def battle(player, monster):
    print(f"{player.name}와(과) {monster.name}의 전투를 시작합니다!")
    
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)  # 플레이어가 몬스터를 공격
        if monster.is_alive():
            monster.attack_target(player)  # 몬스터가 플레이어를 공격
        
    if player.is_alive():
        print("전투 승리!")
        exp_gained = monster.level * 20
        player.gain_experience(exp_gained)
    else:
        print("전투 패배..")

# 예시
player = Player("용사")
monster = Monster("오크", 3)

battle(player, monster)
```

```python
import time  

def battle(player, monster):
    print(f"{player.name}와(과) {monster.name}의 전투를 시작합니다!")
    time.sleep(1)  # 1초 대기
    
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)  # 플레이어가 몬스터를 공격
        time.sleep(1)  # 1초 대기
        if monster.is_alive():
            monster.attack_target(player)  # 몬스터가 플레이어를 공격
            time.sleep(1)  # 1초 대기
        
    if player.is_alive():
        print("전투 승리!")
        exp_gained = monster.level * 20
        player.gain_experience(exp_gained)
    else:
        print("전투 패배..")

# main 함수
def main():
    
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    
    # 사용자로부터 플레이어 이름을 입력받아 Player 인스턴스 생성
    player_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(player_name)
    time.sleep(1)  # 1초 대기
    
    # 몬스터와 차례로 전투 진행
    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        time.sleep(1)  # 1초 대기
        battle(player, monster)
        
        # 전투 도중 플레이어가 사망하면 게임 종료
        if not player.is_alive():
            print("게임오버")
            break
    else:
        print("모든 몬스터와의 전투에서 승리했습니다!")

# 게임 실행
if __name__ == "__main__":
    main()
```
